# Imports

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import plotly.express as px
```

# Input and Preprocessing

```python
submission = pd.read_csv('sample_submission.csv')
test = pd.read_csv('Test.csv')
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
test.drop(columns=['Index'], inplace=True)
```

```python
test.head()
```

```python
#Encoding A
label_encoder = preprocessing.LabelEncoder()
df['A'] = label_encoder.fit_transform(df['A'])
```

```python
test['A'] = label_encoder.transform(test['A'])
```

```python
df.head()
```

```python
data = df[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']]
```

```python
target = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=10)
```

# Trying Different Models for Regression

## 1) Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
mean_squared_error(y_test, lr.predict(X_test), squared=False)
```

```python
submission['Target'] = lr.fit(data, target).predict(test)
submission.to_csv('SubmissionLinearRegression.csv', index=False)
```

**SCORE: 2.28850**

## 2) Polynomial Regression

```python
sns.pairplot(pd.read_csv('Train.csv'))
```

We can see squared dependency of Target on all the inputs (except D)!

```python
dfpr = pd.read_csv('Train.csv')
dfpr['A'] = label_encoder.transform(dfpr['A'])
```

```python
dfpr[['Bs','Cs','Es','Fs','Gs','Hs']] = np.square(df[['B','C','E','F','G','H']])
```

```python
dfpr.head()
```

```python
dfpr.drop(columns=['Target'], inplace=True)
```

```python
X_trainp, X_testp, y_trainp, y_testp = train_test_split(dfpr, target, test_size=0.3, random_state=0)
```

```python
pr = LinearRegression()
```

```python
pr.fit(X_trainp, y_trainp)
mean_squared_error(y_testp, pr.predict(X_testp), squared=False)
```

```python
#Importing Testing set and applying same transformation on it then predicting and saving
testp = pd.read_csv('Test.csv')
testp['A'] = label_encoder.transform(testp['A'])
```

```python
testp[['Bs','Cs','Es','Fs','Gs','Hs']] = np.square(testp[['B','C','E','F','G','H']])
```

```python
testp.drop(columns=['Index'], inplace=True)
```

```python
submission['Target'] = pr.fit(dfpr, target).predict(testp)
submission.to_csv('SubmissionPolynomialRegression.csv', index=False)
```

**SCORE: 2.30877 :(**

### Polynomial with Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
rrhp = GridSearchCV(ridge, {'alpha':np.arange(0.001,0.5,0.0005)}, cv=5, scoring='neg_mean_squared_error')
rrhp.fit(dfpr, target)
```

```python
rrhp.best_params_
```

```python
#{'alpha': 0.0045000000000000005}
submission['Target'] = Ridge(**rrhp.best_params_).fit(dfpr, target).predict(testp)
submission.to_csv('SubmissionPolynomialRegressionRidge.csv', index=False)
```

**SCORE: 2.30723**

### Polynomial with Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
lrhp = GridSearchCV(lasso, {'alpha':np.arange(0.001, 0.5, 0.0005)}, cv=5, scoring='neg_mean_squared_error')
lrhp.fit(dfpr, target)
```

```python
lrhp.best_params_
```

```python
#{'alpha': 0.001}
mean_squared_error(y_testp, Lasso(**lrhp.best_params_).fit(X_trainp, y_trainp).predict(X_testp), squared=False)
```

```python
lasso = Lasso(**lrhp.best_params_)
submission['Target'] = lasso.fit(dfpr, target).predict(testp)
submission.to_csv('SubmissionPolynomialRegressionLasso002.csv', index=False)
```

**SCORE: 2.28067**

## 3) Gradient Boosting

### CatBoost

```python
from catboost import CatBoostRegressor
```

```python
catboost = CatBoostRegressor()
```

```python
param_grid = {
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'depth': [4, 5, 6, 7, 8, 9, 10],
    'l2_leaf_reg': [1, 3, 5, 7],
    'iterations': [500, 1000, 1500],
    'early_stopping_rounds': [50, 100, 200],
    'eval_metric': ['RMSE']
}
cbhp = GridSearchCV(catboost, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, pre_dispatch='2*n_jobs')
cbhp.fit(data, target, verbose=False)
```

```python
#The above algorithm took like 4 hours to run
cbhp.best_params_
```

```python
#Saving the above output to avoid rerunning
cbhpbest= {'depth': 8,
 'early_stopping_rounds': 50,
 'eval_metric': 'RMSE',
 'iterations': 500,
 'l2_leaf_reg': 7,
 'learning_rate': 0.05}
```

```python
mean_squared_error(y_test, CatBoostRegressor(**cbhpbest).fit(X_train, y_train, verbose=False).predict(X_test), squared=False)
```

```python
submission['Target'] = CatBoostRegressor(**cbhpbest).fit(data, target, verbose=False).predict(test)
submission.to_csv('SubmissionCatBoostHP.csv', index=False)
```

**SCORE: 2.21397**

## 5) K-Nearest Regressor

This was the most promising algorithm in my last run

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Finding best hyperparameters
```

```python
knr = KNeighborsRegressor()
knrhp = GridSearchCV(knr, {'weights': ['uniform', 'distance'],
                           'p': [1,2], 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']},
                         cv=10, scoring='neg_mean_squared_error')
knrhp.fit(X_train, y_train)
knrhp.best_params_
```

```python
knr_rmse = []
for k in range(1,50):
    knr = KNeighborsRegressor(algorithm='brute', n_neighbors=k, p=2, weights='distance')
    knr.fit(X_train, y_train)
    error = (mean_squared_error(y_test, knr.predict(X_test), squared=False))
    knr_rmse.append(error)
```

```python
plt.plot(np.arange(1,50), knr_rmse)
```

```python
knr = KNeighborsRegressor(algorithm='brute', p=2, weights='distance', n_neighbors=22)
submission['Target'] = knr.fit(data, target).predict(test)
submission.to_csv('SubmissionKNeighbors.csv', index=False)
```

**SCORE: 2.19561**

# Hyperparameter Tuning of MLPRegressor

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp2hpbest = mlp2hp.best_params_
print(mlp2hpbest)
mean_squared_error(y_test, MLPRegressor(**mlp2hpbest).fit(X_train, y_train).predict(X_test), squared=False)
```

```python
#{'activation': 'tanh', 'alpha': 0.15, 'hidden_layer_sizes': (10,), 'solver': 'lbfgs'}
submission['Target'] = MLPRegressor(**mlp2hpbest).fit(data, target).predict(test)
submission.to_csv('SubmissionMLP2HP.csv', index=False)
```

**SCORE: 2.11347!**

```python
#Playing with alpha a bit more
mlp3 = MLPRegressor(activation= 'tanh', hidden_layer_sizes= (10,), solver='lbfgs')
param_grid={
    'alpha' : np.arange(0.001,0.200,0.001),
    'activation': ['tanh'],
    'hidden_layer_sizes': (10,), 
    'solver': ['lbfgs']
}
mlp3hp = GridSearchCV(mlp3, param_grid, cv=4, scoring='neg_mean_squared_error', n_jobs=-1, pre_dispatch='2*n_jobs')
mlp3hp.fit(data, target)
```

```python
mlp3hpbest = mlp3hp.best_params_
print(mlp3hpbest)
mean_squared_error(y_test, MLPRegressor(**mlp3hpbest).fit(X_train, y_train).predict(X_test), squared=False)
```

```python
#{'activation': 'tanh', 'alpha': 0.004, 'hidden_layer_sizes': (10,), 'solver': 'lbfgs'}
submission['Target'] = MLPRegressor(**mlp3hpbest).fit(data, target).predict(test)
submission.to_csv('SubmissionMLP3HP.csv', index=False)
```
