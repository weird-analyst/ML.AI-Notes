# Imports

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import plotly.express as px
```

# Input and Preprocessing

```python
submission = pd.read_csv('sample_submission.csv')
test = pd.read_csv('Test.csv')
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
test.drop(columns=['Index'], inplace=True)
```

```python
test.head()
```

```python
#Encoding A
label_encoder = preprocessing.LabelEncoder()
df['A'] = label_encoder.fit_transform(df['A'])
```

```python
test['A'] = label_encoder.transform(test['A'])
```

```python
df.head()
```

```python
data = df[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']]
```

```python
target = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=10)
```

# Trying Different Models for Regression

## 1) Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
mean_squared_error(y_test, lr.predict(X_test), squared=False)
```

```python
submission['Target'] = lr.fit(data, target).predict(test)
submission.to_csv('SubmissionLinearRegression.csv', index=False)
```

**SCORE: 2.28850**

## 2) Polynomial Regression

```python
sns.pairplot(pd.read_csv('Train.csv'))
```

We can see squared dependency of Target on all the inputs (except D)!

```python
dfpr = pd.read_csv('Train.csv')
dfpr['A'] = label_encoder.transform(dfpr['A'])
```

```python
dfpr[['Bs','Cs','Es','Fs','Gs','Hs']] = np.square(df[['B','C','E','F','G','H']])
```

```python
dfpr.head()
```

```python
dfpr.drop(columns=['Target'], inplace=True)
```

```python
X_trainp, X_testp, y_train, y_test = train_test_split(dfpr, target, test_size=0.3, random_state=0)
```

```python
pr = LinearRegression()
```

```python
pr.fit(X_trainp, y_train)
mean_squared_error(y_test, pr.predict(X_testp), squared=False)
```

```python
#Importing Testing set and applying same transformation on it then predicting and saving
testp = pd.read_csv('Test.csv')
testp['A'] = label_encoder.transform(testp['A'])
```

```python
testp[['Bs','Cs','Es','Fs','Gs','Hs']] = np.square(testp[['B','C','E','F','G','H']])
```

```python
testp.drop(columns=['Index'], inplace=True)
```

```python
submission['Target'] = pr.fit(dfpr, target).predict(testp)
submission.to_csv('SubmissionPolynomialRegression.csv', index=False)
```

**SCORE: 2.30877 :(**

## 3) Gradient Boosting

### CatBoost

```python
from catboost import CatBoostRegressor
```

```python
catboost = CatBoostRegressor()
```

```python
param_grid = {
    'learning_rate': np.arange(0.01, 0.50, 0.01),
    'depth': [4, 6, 8, 10],
    'l2_leaf_reg': [1, 3, 5, 7],
    'bagging_temperature': [0.8, 1.0, 1.2],
    'border_count': [32, 64, 128],
    'iterations': [500, 1000, 1500],
    'early_stopping_rounds': [50, 100, 200],
    'eval_metric': ['RMSE']
}
cbhp = GridSearchCV(catboost, param_grid, cv=5, scoring='neg_mean_squared_error')
cbhp.fit(data, target, verbose=False)
```

```python
mean_squared_error(y_test, CatBoostRegressor(**cbhp.best_params_).fit(X_train, y_train).predict(X_test), squared=False)
```

```python
submission['Target'] = CatBoostRegressor(**cbhp.best_params_).fit(data, target, verbose=False).predict(test)
submission.to_csv('SubmissionCatBoostHP.csv', index=False)
```

**SCORE:**

### XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor()
param_grid = {
    'learning_rate': [0.01, 0.1, 0.3],
    'max_depth': [3,4,5,6,7],
    'min_child_weight': [1, 3, 5],
    'gamma': [0.0, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'reg_alpha': [0.0, 0.1, 0.5],
    'reg_lambda': [0.0, 0.1, 0.5],
    'n_estimators': [100, 500, 1000],
    'eval_metric': ['rmse']
}
xgbhp = GridSearchCV(xgb, param_grid, cv=5, scoring='neg_mean_squared_error')
xgbhp.fit(data, target, verbose=False)
```

```python
submission['Target'] = XGBRegressor(**xgbhp.best_params_).fit(data, target, verbose=False).predict(test)
submission.to_csv('SubmissionXGBoostHP.csv', index=False)
```

## 5) K-Nearest Regressor

This was the most promising algorithm in my last run

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Finding best hyperparameters
```

```python
knr = KNeighborsRegressor()
knrhp = GridSearchCV(knr, {'weights': ['uniform', 'distance'],
                           'p': [1,2], 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']},
                         cv=10, scoring='neg_mean_squared_error')
knrhp.fit(X_train, y_train)
knrhp.best_params_
```

```python
knr_rmse = []
for k in range(1,50):
    knr = KNeighborsRegressor(algorithm='brute', n_neighbors=k, p=2, weights='distance')
    knr.fit(X_train, y_train)
    error = (mean_squared_error(y_test, knr.predict(X_test), squared=False))
    knr_rmse.append(error)
```

```python
px.line(x=np.arange(1,50), y=knr_rmse)
```

```python
knr = KNeighborsRegressor(algorithm='brute', p=2, weights='distance', n_neighbors=22)
submission['Target'] = knr.fit(data, target).predict(test)
submission.to_csv('SubmissionKNeighbors.csv', index=False)
```

**SCORE: 2.19561** (best till now)
